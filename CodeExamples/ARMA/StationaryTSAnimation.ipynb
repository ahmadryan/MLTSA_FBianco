```
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time


# create figure and axes 
fig, (ax1, ax2) = plt.subplots(2, figsize=(20, 10))
N = 2000
np.random.seed(123)
y = np.random.randn(N)
t = np.linspace(0, N, N, endpoint=False)


# creating our line objects for the plots
ts, = ax1.plot(t, y, '-k', lw=2)
dist = ax2.hist(y[0:500])
plt.show()
def gauss(x, m, s):
  return np.exp(-(x-m)**2 / s**2)

for i in range(0,1500,20):
  fig, (ax1, ax2) = plt.subplots(2, figsize=(20, 10))
  clear_output(wait=True)
  x = np.arange(-3,3,0.1)
  ts, = ax1.plot(t, y, '-k', lw=2)
  ax1.plot(t[i:i+500], y[i:i+500], '-r', lw=2)
  dist = ax2.hist(y[i:i+500], bins=20)
  ax2.plot(x, gauss(x, 0, 1) * 70)
  plt.show() 
  ax2.set_xlim(-4,4)
  ax2.set_ylim(0,75)
  time.sleep(0.3)
```
